```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the data
data = {
    "Classifier": ["Logistic regression", "KNN", "Random forest", "Decision tree", "PCA"],
    "Precision": [0.95, 0.95, 0.97, 0.97, 0.92],
    "Recall": [0.99, 0.99, 0.98, 0.97, 0.99],
    "F1 Score": [0.97, 0.97, 0.98, 0.97, 0.96],
}

# Create DataFrame
df = pd.DataFrame(data)

# Set up the plot
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Melt the DataFrame to long format for seaborn compatibility
df_melted = df.melt("Classifier", var_name="Metric", value_name="Score")

# Plot the data
sns.barplot(data=df_melted, x="Classifier", y="Score", hue="Metric", palette="muted")

# Customize plot aesthetics
plt.title("Model Performance Metrics for Non-Diabetes Prediction")
plt.ylim(0.5, 1.0)
plt.xlabel("Classifier")
plt.ylabel("Score")
plt.legend(title="Metric", loc="upper left")

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the data
data = {
    "Classifier": ["Logistic regression", "KNN", "Random forest", "Decision tree", "PCA"],
    "Precision": [0.84, 0.77, 0.78, 0.69, 0.54],
    "Recall": [0.43, 0.45, 0.73, 0.74, 0.14],
    "F1 Score": [0.57, 0.57, 0.76, 0.71, 0.22],
}

# Create DataFrame
df = pd.DataFrame(data)

# Set up the plot
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Melt the DataFrame to long format for seaborn compatibility
df_melted = df.melt("Classifier", var_name="Metric", value_name="Score")

# Plot the data
sns.barplot(data=df_melted, x="Classifier", y="Score", hue="Metric", palette="muted")

# Customize plot aesthetics
plt.title("Model Performance Metrics for Diabetes Prediction")
plt.ylim(0, 1.0)
plt.xlabel("Classifier")
plt.ylabel("Score")
plt.legend(title="Metric", loc="upper left")

# Display the plot
plt.show()
```
